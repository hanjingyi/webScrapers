```python
import nltk
from nltk import sent_tokenize
```

```python
import itertools
```

# Read scrapt file and tokenize the paragraph to sentences.

```python
def readfile(file):
    with open(file,'r',encoding='utf8') as f:
        l = []
        for line in f:
            try:
                l.append(line.strip().split('"')[3])
            except IndexError:
                pass
        return l
```

```python
readfile('SCIELO_extraction.json')
```

```python
def sent_tok(file):
    l = readfile(file)
    tok = [sent_tokenize(line) for line in l]
    sents = list(itertools.chain.from_iterable(tok))
    return sents
    
```
