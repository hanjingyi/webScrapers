
Link collection - level.1


Given the main page "Ciencias de la Salud", return a list of all links of magzines.





```python
# Link collection - level.1
# Given the main page "Ciencias de la Salud", return a list of all links of magzines.

def extract_links_from_mainpage(link):
    req = Request(link)
    web_page = urlopen(req)
    soup = BeautifulSoup(web_page, 'lxml')
    
    links = []
    
    for link in soup.findAll('a', attrs={'href':re.compile('^http'),
                                         'target':'_blank'}):
        links.append(link.get('href'))
    
    return links
```

Link collection - Level.2


Collect "todos" links from the links extracted in level-1

```python
# Link collection - Level.2
# Collect "todos" links from the links extracted in level-1


def extract_level2(links_level1):
    links_level2 = []
    headers={
        'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'
    }
    for idx,link in enumerate(links_level1):
        try:
            req = Request(link, headers=headers)
            web_page = urlopen(req)
            soup = BeautifulSoup(web_page, 'lxml')
            all_herfs = soup.findAll(name='a')
            
            for x in all_herfs:
                img = x.findChild('img')
                if img is not None:
                    if img.get('src') == '/img/pt/all.gif':
                        links_level2.append(x.get('href'))
            
        except HTTPError as ex:
            print("HTTP error with URL: {}".format(link))
        except URLError as ex:
            print("Name resolution error with URL: {}".format(link))
        print("In total there are {} links, now number {} is being processed".format(len(links_level1),idx+1))
    return links_level2
        
```

```python
def scrape_content(scielo_level4):
    list_content=[]
    headers={
        'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'
    }
    for idx,link in enumerate(scielo_level4):
        try:
            req = Request(link, headers=headers)
            web_page = urlopen(req)
            soup = BeautifulSoup(web_page, 'lxml')

            for text in soup.findAll('font'):
                # Get ride of some unicode chars
                clean = re.sub(r'[\xa0\x93\x94\x96\t\x85]'," ",text.get_text())
                list_content.append(clean)
                
        except HTTPError as ex:
            print("HTTP error with: {}".format(link))
        except URLError as ex:
            print("Name resolution error with: {}".format(link))
        except socket.timeout as ex:
            print("RemoteDisconnected error with:{}".format(link))
        except http.client.HTTPException as ex:
            print('HTTPException error with {}'.format(link))
        print("In total there are {} links, now number {} is being process".format(len(scielo_level4),idx+1))
           
     #Remove empty lines         
        output = list(filter(str.strip,list_content))
     #Tokenize paragraph to sentences          
        tok_output = [sent_tokenize(line) for line in output]
        sents = list(itertools.chain.from_iterable(tok_output))
        
        with open('final_content_extraction_4.txt','w',encoding='utf8') as o:
            o.write('\n'.join(line for line in sents))
    
#    return sents
```
